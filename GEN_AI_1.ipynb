```python
import gradio as gr
from transformers import pipeline

# Load sentiment analysis pipeline from Hugging Face
sentiment_analyzer = pipeline("sentiment-analysis")

# Function to predict sentiment
def predict_sentiment(text):
    result = sentiment_analyzer(text)
    return result[0]['label'], result[0]['score']

# Create a Gradio interface
iface = gr.Interface(
    fn=predict_sentiment,                  # function to call
    inputs=gr.Textbox(label="Enter Text"),  # input text box
    outputs=[gr.Textbox(label="Sentiment"), gr.Textbox(label="Confidence Score")],  # output text boxes
    title="Sentiment Analysis",            # title of the app
    description="Enter any text to analyze its sentiment: positive or negative."
)

# Launch the interface
iface.launch()
```

```python

# Import necessary libraries
import gradio as gr
from transformers import pipeline

# Load a pre-trained text classification model from Hugging Face
classifier = pipeline("text-classification", model="SamLowe/roberta-base-go_emotions")

# Define the function for text classification
def classify_text(text):
    results = classifier(text)
    return {result['label']: result['score'] for result in results}

# Create a Gradio interface
description = "Enter a piece of text to classify its emotion or category using a pre-trained model."
app = gr.Interface(
    fn=classify_text,  # Function to call
    inputs="text",     # Input type
    outputs="label",   # Output type (shows labels and scores)
    title="Text Classification App",
    description=description
)

# Launch the app
app.launch()
```

```python
# Import necessary libraries
import gradio as gr
from transformers import pipeline

# Load the summarization pipeline
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")

# Define the summarization function
def summarize(text):
    if len(text.strip()) == 0:
        return "Please provide some text to summarize."
    summary = summarizer(text, max_length=150, min_length=40, do_sample=False)
    return summary[0]['summary_text']

# Create a Gradio interface
interface = gr.Interface(
    fn=summarize,  # Function to wrap
    inputs=gr.Textbox(lines=10, label="Input Text"),  # Input component
    outputs=gr.Textbox(label="Summary"),  # Output component
    title="Text Summarization App",
    description="Enter a long piece of text and get a concise summary using the DistilBART-CNN model."
)

# Launch the interface
interface.launch()
```

```python
# Import necessary libraries
import gradio as gr
from transformers import pipeline

# Load the text generation model
generator = pipeline("text-generation", model="gpt2")

# Define the text generation function
def generate_text(prompt, max_length=50, num_return_sequences=1):
    """
    Generate text based on the input prompt.

    Parameters:
        - prompt: The input text prompt.
        - max_length: Maximum length of the generated text.
        - num_return_sequences: Number of sequences to generate.

    Returns:
        - Generated text.
    """
    if len(prompt.strip()) == 0:
        return "Please provide a valid input prompt."

    try:
        results = generator(prompt, max_length=max_length, num_return_sequences=num_return_sequences)
        return "\n\n".join([result["generated_text"] for result in results])
    except Exception as e:
        return f"An error occurred: {str(e)}"

# Create Gradio interface components
input_prompt = gr.Textbox(
    lines=2,
    label="Input Prompt",
    placeholder="Enter your prompt here..."
)

max_length_slider = gr.Slider(
    minimum=20,
    maximum=200,
    value=50,
    step=10,
    label="Maximum Length"
)

num_sequences_slider = gr.Slider(
    minimum=1,
    maximum=5,
    value=1,
    step=1,
    label="Number of Sequences"
)

output_textbox = gr.Textbox(
    lines=10,
    label="Generated Text",
    interactive=False  # Make the output textbox read-only
)

# Create the Gradio interface
interface = gr.Interface(
    fn=generate_text,  # Function to call
    inputs=[input_prompt, max_length_slider, num_sequences_slider],
    outputs=output_textbox,
    title="Text Generation App",
    description="Generate creative text based on your input prompt using GPT-2.",
)

# Launch the interface
if __name__ == "__main__":
    interface.launch()
```
